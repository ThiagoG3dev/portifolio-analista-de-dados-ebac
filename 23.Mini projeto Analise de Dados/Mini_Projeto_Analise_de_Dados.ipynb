### O que queremos saber/fazer?

1. Valor Total da Folha Salarial -> Qual foi o gasto total com salários de funcionários pela empresa? <br>
    Sugestão: calcule o salário total de cada funcionário, salário + benefícios + impostos, depois some todos os salários
    
    
2. Qual foi o faturamento da empresa?<br>
    Sugestão: calcule o faturamento total de cada serviço e depois some o faturamento de todos
    
    
3. Qual o % de funcionários que já fechou algum contrato?<br>
    Sugestão: na base de serviços temos o funcionário que fechou cada serviço. Mas nem todos os funcionários que a empresa tem já fecharam algum serviço.<br>
    . Na base de funcionários temos uma lista com todos os funcionários<br>
    . Queremos calcular Qtde_Funcionarios_Fecharam_Serviço / Qtde_Funcionários_Totais<br>
    . Para calcular a qtde de funcionários que fecharam algum serviço, use a base de serviços e conte quantos funcionários tem ali. Mas lembre-se, cada funcionário só pode ser contado uma única vez.<br><br>
    Dica: se você aplicar o método .unique() em uma variável que é apenas 1 coluna de um dataframe, ele vai excluir todos os valores duplicados daquela coluna.<br>
    Ex: unicos_colunaA = dataframe['colunaA'].unique() te dá como resposta uma lista com todos os itens da colunaA aparecendo uma única vez. Todos os valores repetidos da colunaA são excluidos da variável unicos_colunaA
    
    
4. Calcule o total de contratos que cada área da empresa já fechou


5. Calcule o total de funcionários por área


6. Qual o ticket médio mensal (faturamento médio mensal) dos contratos?<br>
    Dica: .mean() calcula a média -> exemplo: media_colunaA = dataframe['colunaA'].mean()

Observação Importante: Se o seu código der um erro na hora de importar os arquivos:<br>

Use separador ";" (ponto e vírgula) para resolver

```python
import pandas as pd

# abrindo arquivos csv
funcionarios_df = pd.read_csv("CadastroFuncionarios.csv", sep=';', decimal=',')
clientes_df = pd.read_csv("CadastroClientes.csv", sep=';', decimal=',')

# abrindo arquivo excel
servicos_xlsx = pd.read_excel("BaseServi‡osPrestados.xlsx")
# salvando excel em csv
servicos_xlsx.to_csv("servicos.csv")

# abrindo arquivo excel em csv
servicos_df = pd.read_csv("servicos.csv", sep=',')
```

```python
# Removendo coluna DataFrame servicos
#servicos_df = servicos_df.drop(['Unnamed: 0'],axis=1)

# dataframes

display(funcionarios_df)
display(clientes_df)
display(servicos_df)
```

```python
# 1. Valor Total da Folha Salarial -> Qual foi o gasto total com salários de funcionários pela empresa?

# somando por coluna
total_salario = funcionarios_df['Salario Base'].sum()
total_imposto = funcionarios_df['Impostos'].sum()
total_beneficios = funcionarios_df['Beneficios'].sum()
total_vt = funcionarios_df['VT'].sum()
total_vr = funcionarios_df['VR'].sum()

# soma total
salario_total_funcionarios = total_salario + total_imposto + total_beneficios + total_vt + total_vr
print(f"Total da folha salarial é de R$ {salario_total_funcionarios:,}")
```

```python
# 1. Valor Total da Folha Salarial -> Qual foi o gasto total com salários de funcionários pela empresa?

funcionarios_df['Salario Total'] = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR']
print(f"Total da folha salarial é de R$ {funcionarios_df['Salario Total'].sum():,}")
```

```python
# 2. Qual foi o faturamento da empresa?

# novo data frame
faturamentos_df = servicos_df[['ID Cliente', 'Tempo Total de Contrato (Meses)']].merge(clientes_df[['ID Cliente','Cliente', 'Valor Contrato Mensal']], on='ID Cliente')

# ordenando data frame
faturamentos_df = faturamentos_df[['ID Cliente', 'Cliente', 'Valor Contrato Mensal', 'Tempo Total de Contrato (Meses)']]

# imprimindo data frame
display(faturamentos_df)
```

```python
# coluna de faturamento total
faturamentos_df['Faturamento Total'] = faturamentos_df['Valor Contrato Mensal'] * faturamentos_df['Tempo Total de Contrato (Meses)']

# somando coluna e printando
print(f"Faturamento total {faturamentos_df['Faturamento Total'].sum():2,}")
```

```python
# 3. Qual o % de funcionários que já fechou algum contrato?

# contar quantidade de funcionarios que fecharam contratos, distinct count
qtd_funcionarios_fecharam_contrato = len(servicos_df['ID Funcionário'].unique())

# quantidade de funcionarios total
qtd_funcionario = len(funcionarios_df['ID Funcionário'])

# dividindo percentual e printando
print(f"percentual de quantidade de funcionarios que fecharam contrato {qtd_funcionarios_fecharam_contrato / qtd_funcionario:.2%}")
```

```python
# 4. Calcule o total de contratos que cada área da empresa já fechou

# nova tabela
contratos_area_df = servicos_df[['ID Funcionário','Codigo do Servico']].merge(funcionarios_df[['ID Funcionário' , 'Area']], on='ID Funcionário')

# printa tabela
display(contratos_area_df)
```

```python
# contando quantidade de funcionario por Area
qtd_contratos_area = contratos_area_df['Area'].value_counts()
print(qtd_contratos_area)

qtd_contratos_area.plot(kind='bar')
```

```python
# 5. Calcule o total de funcionários por área

qtd_funcionarios_area = funcionarios_df['Area'].value_counts()
print(qtd_funcionarios_area)

qtd_funcionarios_area.plot(kind='bar')
```

```python
# 6. Qual o ticket médio mensal (faturamento médio mensal) dos contratos?

ticket_medio = clientes_df['Valor Contrato Mensal'].mean()

print(f"Ticket Medio Mensal R$ {ticket_medio:,.2f}")
```
